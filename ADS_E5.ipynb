```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data
data = pd.read_csv("/content/Sleep_Efficiency.csv")

# Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(x='Sleep duration', y='Sleep efficiency', data=data)
plt.title('Line plot of Sleep Duration vs. Sleep Efficiency')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Sleep duration', y='Sleep efficiency', data=data)
plt.title('Scatter plot of Sleep Duration vs. Sleep Efficiency')
plt.show()
```

```python
#Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Sleep duration', y='Sleep efficiency', data=data)
plt.title('Box plot of Sleep Duration vs. Sleep Efficiency')
plt.show()
```

```python

# Histogram
plt.figure(figsize=(10, 6))
plt.hist(data['Sleep duration'], bins=20, alpha=0.7)
plt.xlabel('Sleep Duration')
plt.ylabel('Frequency')
plt.title('Histogram of Sleep Duration')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=data)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# Heatmap for all attributes except ID
plt.figure(figsize=(10, 8))
sns.heatmap(data.drop('ID', axis=1).corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python

# Pie chart for Sleep duration
plt.figure(figsize=(8, 8))
plt.pie(data['Sleep duration'].value_counts(), labels=data['Sleep duration'].unique(), autopct='%1.1f%%')
plt.title('Pie Chart of Sleep duration distribution')
plt.show()
```

```python
# Area plot for Age and Sleep duration
plt.figure(figsize=(10, 6))
sns.lineplot(x='Age', y='Sleep duration', data=data, estimator=None, lw=1)
plt.fill_between(data['Age'], data['Sleep duration'], alpha=0.2)
plt.title('Area plot of Age vs. Sleep Duration')
plt.xlabel('Age')
plt.ylabel('Sleep Duration')
plt.show()
```

```python

# Violin plot for Sleep efficiency by Gender
plt.figure(figsize=(10, 6))
sns.violinplot(x='Gender', y='Sleep efficiency', data=data)
plt.title('Violin plot ofSleep Efficiency by Gender')
plt.show()
```

```python

# Bubble chart for Sleep duration, Sleep efficiency and age
plt.figure(figsize=(10, 8))
sns.scatterplot(data=data, x='Sleep duration', y='Sleep efficiency', size='Age', sizes=(20, 200), hue='Age')
plt.title('Bubble chart of Sleep duration, Sleep efficiency and Age')
plt.xlabel('Sleep duration')
plt.ylabel('Sleep efficiency')
plt.legend(title='Age')
plt.show()
```
